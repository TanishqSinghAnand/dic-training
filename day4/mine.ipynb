```python
plt.figure(figsize=(8, 6))         # 8 inches wide, 6 inches tall
plt.plot(df["Open"][:150], df["Close"][:150])
plt.title('Using plt.figure(figsize=(8,6))')
plt.show()
```

```python
plt.figure(figsize=(8, 6))         # 8 inches wide, 6 inches tall
plt.plot(df["Date"], df["Close"])
plt.title('Using plt.figure(figsize=(8,6))')
plt.show()
```

```python
df["Open"] -= 5
df["Close"] += 5
df["High"] += 2.5
df["Low"] -= 2.5
```

```python
plt.figure(figsize=(13, 10))         # 13 inches wide, 10 inches tall
plt.plot(df["Date"][:100], df["Close"][:100], label='Close', color='red')
plt.plot(df["Date"][:100], df["Open"][:100], label='Open', color='blue')
plt.plot(df["Date"][:100], df["High"][:100], label='High', color='green')
plt.plot(df["Date"][:100], df["Low"][:100], label='Low', color='orange')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()    
plt.title('Stock Prices Over Time')
plt.show()
```
