```python
laptops = {
    'RAM_GB':[4,8,16,32,64,8,16,32,64,128],
    'Storage_GB':[128,256,512,512,1024,256,512,1024,2048,4096],
    'CPU_Speed_GHz':[2.0,2.2,2.5,3.0,3.5,2.1,2.4,2.8,3.2,3.8],
    'Battery_mAh':[3000,3500,4000,4500,5000,3500,4200,4800,5200,6000],
    'Price':[30000,45000,70000,90000,150000,50000,80000,120000,180000,250000]
}
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(laptops)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(11,11))

plt.subplot(2,2,1)
plt.scatter(df['RAM_GB'], df['Storage_GB'], color='blue')
plt.title('RAM vs Storage')
plt.xlabel('RAM (GB)')
plt.ylabel('Storage (GB)')

plt.subplot(2,2,2)
plt.scatter(df['RAM_GB'], df['CPU_Speed_GHz'], color='red')
plt.title('RAM vs CPU Speed')
plt.xlabel('RAM (GB)')
plt.ylabel('CPU Speed (GHz)')

plt.subplot(2,2,3)
plt.scatter(df['RAM_GB'], df['Battery_mAh'], color='green')
plt.title('RAM vs Battery')
plt.xlabel('RAM (GB)')
plt.ylabel('Battery (mAh)')

plt.subplot(2,2,4)
plt.scatter(df['RAM_GB'], df['Price'], color='maroon')
plt.title('RAM vs Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Price')

plt.show()
```

```python
import seaborn as sns
corr_matrix = df.corr()
plt.figure(figsize=(10,7.5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
x = df

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=df['Price'], cmap='viridis', s=100)
plt.colorbar(label='Price')
plt.title('PCA of Laptop Features')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
