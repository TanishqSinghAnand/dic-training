```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

fruits  = ['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry']
calories = [95, 89, 50, 277, 106]
colors   = ['#e74c3c', '#f1c40f', '#c0392b', '#8e44ad', '#27ae60']

# Vertical bars
axes[0].bar(fruits, calories, color=colors, edgecolor='black', width=0.6)
axes[0].set_title('plt.bar() — Vertical')
axes[0].set_ylabel('Calories')
axes[0].tick_params(axis='x', rotation=15)

# Horizontal bars (barh)
axes[1].barh(fruits, calories, color=colors, edgecolor='black', height=0.6)
axes[1].set_title('plt.barh() — Horizontal')
axes[1].set_xlabel('Calories')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
x      = np.arange(1, 7)
y      = np.array([3.5, 4.2, 3.8, 5.1, 4.7, 5.5])
errors = np.array([0.3, 0.5, 0.2, 0.4, 0.6, 0.3])

plt.errorbar(x, y,
             yerr=errors,           # vertical error size
             fmt='o-',              # 'o' dots connected by '-' line
             color='purple',
             ecolor='gray',         # error bar color
             elinewidth=2,
             capsize=6,             # horizontal caps on error bars
             markersize=8,
             lw=2)

plt.title('plt.errorbar() — measurements with uncertainty')
plt.xlabel('Experiment')
plt.ylabel('Result')
plt.grid(True, alpha=0.3)
plt.show()
```

# Final Script

```python
plt.style.use('ggplot')
np.random.seed(42)

fig = plt.figure(figsize=(16, 10))
plt.suptitle('📊 Student Performance Dashboard', fontsize=18, fontweight='bold', y=0.98)

# ── Chart 1: Line — Score over time ──────────────────
ax1 = fig.add_subplot(2, 3, 1)
weeks  = np.arange(1, 13)
scores = 50 + np.cumsum(np.random.randn(12) * 3)
ax1.plot(weeks, scores, color='steelblue', lw=2, marker='o', markersize=5)
ax1.axhline(y=np.mean(scores), color='red', ls='--', lw=1.5, label=f'Avg: {np.mean(scores):.1f}')
ax1.set_title('Score over weeks')
ax1.set_xlabel('Week')
ax1.set_ylabel('Score')
ax1.legend(fontsize=9)
ax1.grid(True, alpha=0.4)

# ── Chart 2: Bar — Subject marks ─────────────────────
ax2 = fig.add_subplot(2, 3, 2)
subjects = ['Maths', 'Science', 'English', 'History', 'Art']
marks    = [88, 74, 91, 65, 82]
bar_colors = ['#2ecc71' if m >= 80 else '#e74c3c' for m in marks]
bars = ax2.bar(subjects, marks, color=bar_colors, edgecolor='black', width=0.6)
for bar, m in zip(bars, marks):
    ax2.text(bar.get_x() + bar.get_width()/2, m + 1, str(m),
             ha='center', fontsize=10, fontweight='bold')
ax2.set_title('Marks by subject (green ≥ 80)')
ax2.set_ylim(0, 105)
ax2.set_ylabel('Marks')

# ── Chart 3: Histogram — Class distribution ──────────
ax3 = fig.add_subplot(2, 3, 3)
class_scores = np.random.normal(72, 12, 200)
ax3.hist(class_scores, bins=25, color='coral', edgecolor='black', alpha=0.85)
ax3.axvline(x=60, color='red',   lw=2, ls='--', label='Pass: 60')
ax3.axvline(x=np.mean(class_scores), color='navy', lw=2, ls=':', label=f'Mean: {np.mean(class_scores):.1f}')
ax3.set_title('Class score distribution')
ax3.set_xlabel('Score')
ax3.set_ylabel('No. of students')
ax3.legend(fontsize=9)

# ── Chart 4: Scatter — Study time vs Score ───────────
ax4 = fig.add_subplot(2, 3, 4)
study  = np.random.uniform(1, 8, 80)
result = study * 9 + np.random.randn(80) * 6
ax4.scatter(study, result, c=result, cmap='RdYlGn', s=60, alpha=0.75, edgecolors='gray', lw=0.4)
ax4.set_title('Study time vs Score')
ax4.set_xlabel('Hours studied')
ax4.set_ylabel('Score')

# ── Chart 5: Pie — Grade distribution ────────────────
ax5 = fig.add_subplot(2, 3, 5)
grades  = ['A (90+)', 'B (75-90)', 'C (60-75)', 'F (<60)']
counts  = [22, 45, 28, 10]
colors5 = ['#2ecc71', '#3498db', '#f39c12', '#e74c3c']
ax5.pie(counts, labels=grades, colors=colors5,
        autopct='%1.0f%%', startangle=90,
        wedgeprops=dict(edgecolor='white', lw=2))
ax5.set_title('Grade distribution')

# ── Chart 6: Boxplot — Semester comparison ────────────
ax6 = fig.add_subplot(2, 3, 6)
sem1 = np.random.normal(65, 12, 80)
sem2 = np.random.normal(72,  9, 80)
sem3 = np.random.normal(78,  8, 80)
ax6.boxplot([sem1, sem2, sem3], labels=['Sem 1', 'Sem 2', 'Sem 3'],
            patch_artist=True,
            boxprops=dict(facecolor='lightyellow', color='navy'),
            medianprops=dict(color='red', lw=2))
ax6.set_title('Score by semester')
ax6.set_ylabel('Score')
ax6.grid(True, axis='y', alpha=0.4)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('student_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print('\n✅ Dashboard saved as student_dashboard.png')
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Graph', fontdict={'fontname': 'sans-serif', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

# Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

**Pie chart**

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

**Pie chart 2**

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```
