```python
sem1=7.95
sem2=8.5
sem3=8.0
cgpa = (sem1 + sem2 + sem3) / 3
print(f"CGPA : {cgpa}")
```

```python
# Grade calculation
if cgpa <= 4:
    grade = 'F'
elif cgpa > 4 and cgpa <= 5:
    grade = 'D'
elif cgpa>5 and cgpa <= 6:
    grade = 'C'
elif cgpa > 6 and cgpa <= 7:
    grade = 'B'
elif cgpa > 7 and cgpa <= 8:
    grade = 'B+'
elif cgpa > 8 and cgpa <= 9:
    grade = 'A'
elif cgpa > 9 and cgpa <=10:
    grade = 'A+'
else:
    grade = 'Invalid CGPA'
print(f"Grade : {grade}")
```
