```python
class University:
    def __init__(self, universityName):
        self.universityName = universityName

    def displayUniversity(self):
        print("University:", self.universityName)


class Department(University):
    def __init__(self, universityName, departmentName):
        super().__init__(universityName)
        self.departmentName = departmentName

    def displayDepartment(self):
        print("Department:", self.departmentName)


class Student(University):
    def __init__(self, universityName, studentId, studentName, course):
        super().__init__(universityName)
        self.studentId = studentId
        self.studentName = studentName
        self.course = course

    def displayStudent(self):
        print("\nStudent Details")
        print("Student ID:", self.studentId)
        print("Name:", self.studentName)
        print("Course:", self.course)


class Staff(University):
    def __init__(self, universityName, staffId, staffName):
        super().__init__(universityName)
        self.staffId = staffId
        self.staffName = staffName

    def displayStaff(self):
        print("\nStaff Details")
        print("Staff ID:", self.staffId)
        print("Name:", self.staffName)


class Teacher(Staff):
    def __init__(self, universityName, staffId, staffName, subject):
        super().__init__(universityName, staffId, staffName)
        self.subject = subject

    def displayTeacher(self):
        self.displayStaff()
        print("Subject:", self.subject)


class Financial(Staff):
    def __init__(self, universityName, staffId, staffName, designation):
        super().__init__(universityName, staffId, staffName)
        self.designation = designation

    def displayFinancial(self):
        self.displayStaff()
        print("Designation:", self.designation)


department = Department("UIET", "Computer Science")

student = Student(
    "UIET",
    101,
    "Tanishq Singh Anand",
    "B.Tech CSE"
)

teacher = Teacher(
    "UIET",
    201,
    "Minto Rattan",
    "Calculus"
)

financialOfficer = Financial(
    "UIET",
    301,
    "Tarun",
    "Applied Sciences Financial Officer"
)

# Output
department.displayUniversity()
department.displayDepartment()

student.displayStudent()

teacher.displayTeacher()

financialOfficer.displayFinancial()
```

```python
class University:
    def __init__(self, universityName):
        self.universityName = universityName


class Student(University):
    def __init__(self, universityName, studentId, studentName):
        super().__init__(universityName)
        self.studentId = studentId
        self.studentName = studentName
        self.subjectPointers = {}
        self.sgpa = 0.0
        self.cgpa = 0.0
        self.feeStatus = "Pending"

    def viewSubjectPointers(self):
        print("\nSubject-wise Pointers:")
        for subject, pointer in self.subjectPointers.items():
            print(subject, ":", pointer)

    def viewSGPA(self):
        print("SGPA:", self.sgpa)

    def viewCGPA(self):
        print("CGPA:", self.cgpa)

    def viewFeeStatus(self):
        print("Fee Status:", self.feeStatus)


class Staff(University):
    def __init__(self, universityName, staffId, staffName):
        super().__init__(universityName)
        self.staffId = staffId
        self.staffName = staffName


class Teacher(Staff):
    def __init__(self, universityName, staffId, staffName):
        super().__init__(universityName, staffId, staffName)

    def uploadGrades(self, student, subject, pointer):
        student.subjectPointers[subject] = pointer
        print(f"{subject} pointer uploaded.")

    def updateGrades(self, student, subject, pointer):
        student.subjectPointers[subject] = pointer
        print(f"{subject} pointer updated.")

    def updateSGPA(self, student, sgpa):
        student.sgpa = sgpa

    def updateCGPA(self, student, cgpa):
        student.cgpa = cgpa


class Financial(Staff):
    def __init__(self, universityName, staffId, staffName):
        super().__init__(universityName, staffId, staffName)

    def updateFeeStatus(self, student, status):
        validStatus = ["Pending", "Submitted", "Late"]

        if status in validStatus:
            student.feeStatus = status
            print("Fee status updated.")
        else:
            print("Invalid fee status.")



student1 = Student("UIET", 101, "Tanishq")

teacher1 = Teacher("UIET", 201, "Dr Sharma")

financial1 = Financial("UIET", 301, "Mr Verma")

teacher1.uploadGrades(student1, "Python", 9)
teacher1.uploadGrades(student1, "Mathematics", 8)

teacher1.updateSGPA(student1, 8.5)
teacher1.updateCGPA(student1, 8.3)

financial1.updateFeeStatus(student1, "Submitted")

student1.viewSubjectPointers()
student1.viewSGPA()
student1.viewCGPA()
student1.viewFeeStatus()
```

```python
class University:
    def __init__(self, universityName):
        self.universityName = universityName
    def displayInfo(self):
        print("University Name:", self.universityName)
class Department(University):
    def __init__(self, universityName, departmentName):
        super().__init__(universityName)
        self.departmentName = departmentName
    def displayInfo(self):
        print("University Name:", self.universityName)
        print("Department Name:", self.departmentName)
class Student(University):
    def __init__(self, universityName, studentId, studentName):
        super().__init__(universityName)
        self.studentId = studentId
        self.studentName = studentName
        self.subjects = {}
        self.sgpa = 0
        self.cgpa = 0
        self.feeStatus = "Pending"
    def calculateSGPA(self):
        if len(self.subjects) > 0:
            total = sum(self.subjects.values())
            self.sgpa = total / len(self.subjects)
    def calculateCGPA(self):
        self.cgpa = self.sgpa
    def viewDetails(self):
        self.calculateSGPA()
        self.calculateCGPA()
        print("\nStudent ID:", self.studentId)
        print("Student Name:", self.studentName)
        print("Subjects and Pointers:")
        for subject, marks in self.subjects.items():
            print(subject, ":", marks)
        print("SGPA:", round(self.sgpa, 2))
        print("CGPA:", round(self.cgpa, 2))
        print("Fee Status:", self.feeStatus)
class Staff(University):
    def __init__(self, universityName, staffId, staffName):
        super().__init__(universityName)
        self.staffId = staffId
        self.staffName = staffName
class Teacher(Staff):
    def uploadGrades(self, student, subject, pointer):
        student.subjects[subject] = pointer
        print("Grade uploaded successfully")
class Financial(Staff):
    def updateFeeStatus(self, student, status):
        student.feeStatus = status
        print("Fee status updated")
department = Department("UIET", "Computer Science")
teacher = Teacher("UIET", 1, "Minto Rattan")
financial = Financial("UIET", 2, "Tarun")
students = {}
while True:
    print("Menu :-")
    print("1. Add Student")
    print("2. Search Student")
    print("3. Display All Students")
    print("4. Upload Grades")
    print("5. Update Fee Status")
    print("6. View Student Details")
    print("7. University Information")
    print("0. Exit")
    choice = input("Enter choice: ")
    if choice == "1":
        sid = int(input("Enter student id: "))
        name = input("Enter student name: ")
        students[sid] = Student("UIET", sid, name)
        print("Student added")
    elif choice == "2":
        sid = int(input("Enter student id: "))
        if sid in students:
            students[sid].viewDetails()
        else:
            print("Student not found")
    elif choice == "3":
        for student in students.values():
            print(student.studentId, "-", student.studentName)
    elif choice == "4":
        sid = int(input("Enter student id: "))
        if sid in students:
            subject = input("Enter subject: ")
            pointer = float(input("Enter pointer: "))
            teacher.uploadGrades(
                students[sid],
                subject,
                pointer
            )
    elif choice == "5":
        sid = int(input("Enter student id: "))
        if sid in students:
            status = input(
                "Enter status (Pending/Submitted/Late): "
            )
            financial.updateFeeStatus(
                students[sid],
                status
            )
    elif choice == "6":
        sid = int(input("Enter student id: "))
        if sid in students:
            students[sid].viewDetails()
    elif choice == "7":
        department.displayInfo()
    elif choice == "0":
        break
    else:
        print("Invalid choice")
```
