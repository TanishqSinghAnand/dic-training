# THE NETFLIX DATA 


##  Chapter 6: Null Value Handling 


```python
nt_df.isna().sum()
```

```python
null_pct = (nt_df.isna().sum() / len(nt_df) * 100).round(2)
null_pct.sort_values(ascending=False)
```

```python
nt_df['country'] = nt_df['country'].fillna('Unknown')
nt_df['director'] = nt_df['director'].fillna('Not Listed')
print('Nulls remaining in country:', nt_df['country'].isna().sum())
print('Nulls remaining in director:', nt_df['director'].isna().sum())
```

```python
before = len(nt_df)
nt_df = nt_df.dropna(subset=['title'])
after = len(nt_df)
print(f'Rows before: {before} | Rows after dropping missing titles: {after}')
```


##  Chapter 7: Advanced String & Categorical Manipulation 

```python
love_titles = nt_df[nt_df['title'].str.contains('Love', case=False, na=False)]
print(f'Titles containing "Love": {len(love_titles)}')
love_titles['title'].head(10)
```

```python
nt_df['title_upper'] = nt_df['title'].str.upper()
nt_df[['title', 'title_upper']].head(5)
```

```python
nt_df['genres_list'] = nt_df['listed_in'].str.split(', ')
nt_df[['title', 'listed_in', 'genres_list']].head(5)
```

```python
all_genres = nt_df['listed_in'].str.split(', ').explode()
genre_counts = all_genres.value_counts()
print('Top 10 Netflix Genres:')
genre_counts.head(10)
```

```python
nt_df['title_length'] = nt_df['title'].str.len()
print('Longest title:')
print(nt_df.loc[nt_df['title_length'].idxmax(), 'title'])
print('\nShortest title:')
print(nt_df.loc[nt_df['title_length'].idxmin(), 'title'])
```


##  Chapter 8: Grouping & Aggregations


```python
nt_df.groupby('type')['title'].count()
```

```python
nt_df.groupby('type').agg(
    total_titles=('title', 'count'),
    avg_release_year=('release_year', 'mean'),
    earliest_release=('release_year', 'min'),
    latest_release=('release_year', 'max')
).round(1)
```

```python
rating_counts = nt_df['rating'].value_counts()
print('Titles by Rating:')
rating_counts
```

```python
top_countries = nt_df[nt_df['country'] != 'Unknown'].groupby('country')['title'].count()
top_countries_sorted = top_countries.sort_values(ascending=False)
print('Top 10 Content-Producing Countries:')
top_countries_sorted.head(10)
```


##  Chapter 9: Working with Time Series 


```python
print('Before conversion:', nt_df['date_added'].dtype)
nt_df['date_added'] = pd.to_datetime(nt_df['date_added'], errors='coerce')
print('After conversion: ', nt_df['date_added'].dtype)
```

```python
nt_df['year_added'] = nt_df['date_added'].dt.year
nt_df['month_added'] = nt_df['date_added'].dt.month
nt_df['month_name'] = nt_df['date_added'].dt.month_name()
nt_df[['title', 'date_added', 'year_added', 'month_added', 'month_name']].dropna().head(6)
```

```python
nt_df_sorted = nt_df.sort_values(by='date_added', ascending=False)
print('Most recently added titles:')
nt_df_sorted[['title', 'type', 'date_added', 'country']].dropna(subset=['date_added']).head(8)
```
