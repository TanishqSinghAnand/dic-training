```python
class BankAccount:
    bank ="SBI"
    
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
        
    def deposit(self,amount):
        self.balance += amount
        print(f"Deposited : {amount}")
        
    def withdraw(self,amount):
         self.balance -= amount if amount <=self.balance  else print("Insufficient funds!") 
         
    @classmethod
    def bank_name(cls):
        return cls.bank
```

```python
acc = BankAccount("Aryan", 87)
acc.withdraw(87)
acc.deposit(87)
acc.bank_name()
```

```python
# Factorial of a number using recursion
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
```
