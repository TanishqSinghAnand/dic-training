```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
plt.figure(figsize=(8, 4))         # 8 inches wide, 4 inches tall
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Using plt.figure(figsize=(8,4))')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))   # 1 row, 2 columns

axes[0].plot([1, 2, 3], [1, 4, 9], color='blue')
axes[0].set_title('Left chart')

axes[1].bar(['A', 'B', 'C'], [5, 3, 7], color='orange')
axes[1].set_title('Right chart')

plt.show()
```
