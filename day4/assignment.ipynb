```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Pokemon.csv")

print("dataset loaded")

attack_arr = np.array(df["Attack"])

print("\nshape =", attack_arr.shape)
print("mean =", np.mean(attack_arr))
print("median =", np.median(attack_arr))
print("max value =", np.max(attack_arr))
print("min value =", np.min(attack_arr))
print("std dev =", np.std(attack_arr))

print("\nfirst 5 rows")
print(df.head())

print("\nlast 5 rows")
print(df.tail())

print("\ndataset info")
print(df.info())

print("\nsummary stastics")
print(df.describe())

print("\nmissing values")
print(df.isnull().sum())

print("\nselected columns")
print(df[["Name","Attack","Defense"]])

print("\nfiltered rows")
print(df[df["Attack"] > 100])

print("\nsorted data")
print(df.sort_values(by="Attack", ascending=False))

df["Power_Index"] = df["Attack"] + df["Defense"]

print("\nnew column added")
print(df.head())
```

```python
plt.figure(figsize=(8,5))
plt.plot(df.index[:50], df["Attack"][:50], label="attack")
plt.xlabel("pokemon number")
plt.ylabel("attack")
plt.title("attack values of first 50 pokemon")
plt.legend()
plt.grid()
plt.show()

top10 = df.sort_values(by="Attack", ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10["Name"], top10["Attack"])
plt.xlabel("pokemon")
plt.ylabel("attack")
plt.title("top 10 pokemon by attack")
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df["HP"], bins=15)
plt.xlabel("hp")
plt.ylabel("frequency")
plt.title("hp distributon")
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(df["Attack"], df["Defense"])
plt.xlabel("attack")
plt.ylabel("defense")
plt.title("attack vs defense")
plt.grid()
plt.show()

type_count = df["Type 1"].value_counts().head(5)

plt.figure(figsize=(7,7))
plt.pie(type_count,
        labels=type_count.index,
        autopct="%1.1f%%")
plt.title("top 5 pokemon types")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,
                x="Attack",
                y="Defense",
                hue="Legendary")
plt.title("attack vs defense")
plt.show()

plt.figure(figsize=(8,5))
sns.lineplot(data=df.head(50),
             x=df.head(50).index,
             y="Attack")
plt.title("attack trend")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=df,
            x="Generation",
            y="Attack")
plt.title("average attack by generation")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=df,
              x="Generation")
plt.title("pokemon count in each generation")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(data=df,
             x="HP",
             bins=15)
plt.title("hp distributon")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=df,
            x="Legendary",
            y="Attack")
plt.title("attack values")
plt.show()

plt.figure(figsize=(8,5))
sns.violinplot(data=df,
               x="Legendary",
               y="Attack")
plt.title("attack density")
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    annot=True
)
plt.title("correlation heatmap")
plt.show()

sns.pairplot(
    df[["HP",
        "Attack",
        "Defense",
        "Speed",
        "Legendary"]],
    hue="Legendary"
)

plt.show()
```
